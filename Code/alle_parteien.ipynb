```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
erst_17 = pd.read_csv('../../../Daten/erststimmen_aufbereitet.csv')
zweit_17 = pd.read_csv('../../../Daten/zweitstimmen_aufbereitet.csv')
erst_13 = pd.read_csv('../../../Daten/erststimmen_13.csv', index_col = 0)
zweit_13 = pd.read_csv('../../../Daten/zweitstimmen_13.csv', index_col = 0)

wahl_diff = pd.read_csv('../../../Daten/bt_wahl_13/gesamt.csv', index_col = 0)

wahlbezirke = gpd.read_file('../../../Daten/wahlbezirke.shp')
stadtteile = gpd.read_file('../../../Daten/stadtteile.shp')

wahlbezirke = wahlbezirke.rename(columns = {'Gebietsnum': 'Gebietsnummer'})
```

```python
plot_data_13 = wahlbezirke.set_index('Gebietsnummer').join(wahl_diff.set_index('Gebietsnummer'))
plot_data_17_erst = wahlbezirke.set_index('Gebietsnummer').join(erst_17.set_index('Gebietsnummer'))
plot_data_17_zweit = wahlbezirke.set_index('Gebietsnummer').join(zweit_17.set_index('Gebietsnummer'))
```

## CSU

### 2017

Erststimme

```python
fig, ax = plt.subplots()
plot_data_17_erst.plot(ax = ax, column = 'CSU', legend = True, edgecolor = 'grey')
stadtteile.plot(ax = ax, edgecolor = 'Black', facecolor = 'None')
plt.show()
```

Zweitstimme

```python
fig, ax = plt.subplots()
plot_data_17_zweit.plot(ax = ax, column = 'CSU', legend = True, edgecolor = 'grey')
stadtteile.plot(ax = ax, edgecolor = 'Black', facecolor = 'None')
plt.show()
```
