```python
import pandas as pd
import numpy as np
```

```python
# importiere Parteien
csu = pd.read_csv('../../Daten/bt_wahl_13/csu.txt', sep = ';')
spd = pd.read_csv('../../Daten/bt_wahl_13/erst_und_zweit.txt', sep = ';')
fdp = pd.read_csv('../../Daten/bt_wahl_13/fdp.txt', sep = ';')
grüne = pd.read_csv('../../Daten/bt_wahl_13/grüne.txt', sep = ';')
linke = pd.read_csv('../../Daten/bt_wahl_13/linke.txt', sep = ';')
afd = pd.read_csv('../../Daten/bt_wahl_13/afd.txt', sep = ';')
afd = afd.drop(columns = 'AfD_erst')
```

```python
gesamt = csu
gesamt = gesamt.merge(spd, on = 'Name')
gesamt = gesamt.merge(grüne, on = 'Name')
gesamt = gesamt.merge(fdp, on = 'Name')
gesamt = gesamt.merge(linke, on = 'Name')
gesamt = gesamt.merge(afd, on = 'Name')
gesamt = gesamt.rename(columns = {'Name': 'Gebietsnummer'})
```

```python
erst_17 = pd.read_csv('../../Daten/erststimmen_aufbereitet.csv')
zweit_17 = pd.read_csv('../../Daten/zweitstimmen_aufbereitet.csv')
gesamt['Diff_erst_CSU'] = erst_17['CSU'] - gesamt['CSU_erst'] 
gesamt['Diff_zweit_CSU'] = zweit_17['CSU'] - gesamt['CSU_zweit']
gesamt['Diff_erst_SPD'] = erst_17['SPD'] - gesamt['SPD_erst']
gesamt['Diff_zweit_SPD'] = zweit_17['SPD'] - gesamt['SPD_zweit']
gesamt['Diff_erst_GRÜNE'] = erst_17['GRÜNE'] - gesamt['GRÜNE_erst']
gesamt['Diff_zweit_GRÜNE'] = zweit_17['GRÜNE'] - gesamt['GRÜNE_zweit']
gesamt['Diff_erst_FDP'] = erst_17['FDP'] - gesamt['FDP_erst']
gesamt['Diff_zweit_FDP'] = zweit_17['FDP'] - gesamt['FDP_zweit']
gesamt['Diff_erst_DIE LINKE'] = erst_17['DIE LINKE'] - gesamt['DIE LINKE_erst']
gesamt['Diff_zweit_DIE LINKE'] = zweit_17['DIE LINKE'] - gesamt['DIE LINKE_zweit']
gesamt['Diff_erst_AfD'] = erst_17['AfD']
gesamt['Diff_zweit_AfD'] = zweit_17['AfD'] - gesamt['AfD_zweit']
```

```python
maxvorstadt = pd.read_csv('../../Daten/Stadtbezirke/maxvorstadt.txt', sep = ';')
schwabing_west = pd.read_csv('../../Daten/Stadtbezirke/schwabing_west.txt', sep = ';')
moosach = pd.read_csv('../../Daten/Stadtbezirke/moosach.txt', sep = ';')
milbertshofen = pd.read_csv('../../Daten/Stadtbezirke/milbertshofen_amhart.txt', sep = ';')
freimann = pd.read_csv('../../Daten/Stadtbezirke/schwabing_freimann.txt', sep = ';')
hasenbergl = pd.read_csv('../../Daten/Stadtbezirke/feldmoching_hasenbergl.txt', sep = ';')
```

```python
gesamt['Stadtbezirk'] = 'Maxvorstadt'
gesamt.loc[gesamt['Gebietsnummer'].isin(np.array(schwabing_west.Name)), 'Stadtbezirk'] = 'Schwabing-West'
gesamt.loc[gesamt['Gebietsnummer'].isin(np.array(moosach.Name)), 'Stadtbezirk'] = 'Moosach'
gesamt.loc[gesamt['Gebietsnummer'].isin(np.array(milbertshofen.Name)), 'Stadtbezirk'] = 'Milbertshofen'
gesamt.loc[gesamt['Gebietsnummer'].isin(np.array(freimann.Name)), 'Stadtbezirk'] = 'Freimann'
gesamt.loc[gesamt['Gebietsnummer'].isin(np.array(hasenbergl.Name)), 'Stadtbezirk'] = 'Hasenbergl'
```

```python
diff = gesamt.drop(columns = ['CSU_erst', 'SPD_erst', 'GRÜNE_erst', 'FDP_erst', 'DIE LINKE_erst',
                             'CSU_zweit', 'SPD_zweit', 'GRÜNE_zweit', 'FDP_zweit', 'DIE LINKE_zweit', 'AfD_zweit'])
diff = diff.rename(columns = {'Diff_erst_CSU': 'Differenz Erststimme CSU', 'Diff_erst_SPD': 'Differenz Erststimme SPD', 
                      'Diff_erst_GRÜNE': 'Differenz Erststimme GRÜNE', 'Diff_erst_FDP': 'Differenz Erststimme FDP',
                       'Diff_erst_DIE LINKE': 'Differenz Erststimme DIE LINKE', 'Diff_erst_AfD': 'Differenz Erststimme AfD'})
diff = diff.rename(columns = {'Diff_zweit_CSU': 'Differenz Zweitstimme CSU', 'Diff_zweit_SPD': 'Differenz Zweitstimme SPD', 
                      'Diff_zweit_GRÜNE': 'Differenz Zweitstimme GRÜNE', 'Diff_zweit_FDP': 'Differenz Zweitstimme FDP',
                       'Diff_zweit_DIE LINKE': 'Differenz Zweitstimme DIE LINKE', 'Diff_zweit_AfD': 
                       'Differenz Zweitstimme AfD'})
```

```python
diff.head()
```

```python
Erststimme = ['Gebietsnummer', 'Differenz Erststimme CSU', 'Differenz Erststimme SPD', 'Differenz Erststimme GRÜNE', 
             'Differenz Erststimme FDP', 'Differenz Erststimme DIE LINKE', 'Differenz Erststimme AfD']
Zweitstimme = ['Gebietsnummer', 'Differenz Zweitstimme CSU', 'Differenz Zweitstimme SPD', 'Differenz Zweitstimme GRÜNE', 
             'Differenz Zweitstimme FDP', 'Differenz Zweitstimme DIE LINKE', 'Differenz Zweitstimme AfD']
diff_erst = diff[Erststimme]
diff_zweit = diff[Zweitstimme]
```

```python
diff.to_csv('../../Daten/differenzen.csv')
diff_erst.to_csv('../../Daten/differenzen_erststimme.csv')
diff_zweit.to_csv('../../Daten/differenzen_zweitstimme.csv')
```
