Absolute Zahlen

```python
def no_density(column):
    boolean = ('Arbeitslosendichte' not in column) & ('Beschäftigtendichte' not in column) & (
        'Wohndauer' not in column) & ('Empfängerdichte' not in column) & ('Erwerbsfähige' not in column)
    return boolean
    
    

def absolute_values(data):
    output = data[['Gebietsnummer', 'Bevölkerung 2013', 'Bevölkerung 2017', 'Bevölkerung 2020']]
    columns = data.columns[4::]
    for column in columns:
        if '2013' in column:
            if no_density(column):
                new_column = column.replace('Anteil ', '')
                output[new_column] = data['Bevölkerung 2013']*data[column]
            else:  
                new_column = column.replace('Anteil ', '')
                output[new_column] = data[column]
        if '2017' in column:
            if no_density(column):
                new_column = column.replace('Anteil ', '')
                output[new_column] = data['Bevölkerung 2017']*data[column]
            else:  
                new_column = column.replace('Anteil ', '')
                output[new_column] = data[column]
        if '2020' in column:
            if no_density(column):
                new_column = column.replace('Anteil ', '')
                output[new_column] = data['Bevölkerung 2020']*data[column]
            else:  
                new_column = column.replace('Anteil ', '')
                output[new_column] = data[column]
    return output
```

```python
integers = ['Gebietsnummer', 'Bevölkerung 2013', 'Bevölkerung 2017',
       'Bevölkerung 2020', 'Deutsche Bevölkerung 2013',
       'Nichtdeutsche EU Bevölkerung 2013',
       'Nichtdeutsche nicht EU Bevölkerung 2013', 'Deutsche Bevölkerung 2017',
       'Nichtdeutsche EU Bevölkerung 2017',
       'Nichtdeutsche nicht EU Bevölkerung 2017', 'Deutsche Bevölkerung 2020',
       'Nichtdeutsche EU Bevölkerung 2020',
       'Nichtdeutsche nicht EU Bevölkerung 2020',
       'Migrationshintergrund Bevölkerung 2013',
       'Migrationshintergrund Bevölkerung 2017',
       'Migrationshintergrund Bevölkerung 2020', 'Ledige Bevölkerung 2013',
       'Verheiratete Bevölkerung 2013', 'Ledige Bevölkerung 2017',
       'Verheiratete Bevölkerung 2017', 'Ledige Bevölkerung 2020',
       'Verheiratete Bevölkerung 2020', '18 - 24 Jährige Wahlberechtigte 2013',
       '25 - 34 Jährige Wahlberechtigte 2013',
       '35 - 44 Jährige Wahlberechtigte 2013',
       '45 - 59 Jährige Wahlberechtigte 2013',
       '60 Jährige und ältere Wahlberechtigte 2013',
       '18 - 24 Jährige Wahlberechtigte 2017',
       '25 - 34 Jährige Wahlberechtigte 2017',
       '35 - 44 Jährige Wahlberechtigte 2017',
       '45 - 59 Jährige Wahlberechtigte 2017',
       '60 Jährige und ältere Wahlberechtigte 2017',
       '18 - 24 Jährige Wahlberechtigte 2020',
       '25 - 34 Jährige Wahlberechtigte 2020',
       '35 - 44 Jährige Wahlberechtigte 2020',
       '45 - 59 Jährige Wahlberechtigte 2020',
       '60 Jährige und ältere Wahlberechtigte 2020',
       'Einpersonenhaushalte 2013', 'Haushalte mit Kind(ern) 2013',
       'Einpersonenhaushalte 2017', 'Haushalte mit Kind(ern) 2017',
       'Einpersonenhaushalte 2020', 'Haushalte mit Kind(ern) 2020']
```

```python
for column in integers: 
    struktur_absolut[column] = struktur_absolut[column].apply(lambda x: round(x))
```

```python
def differences(data):
    output = pd.DataFrame()
    output['Gebietsnummer'] = data['Gebietsnummer']
    for column in data.columns:
        if '2013' in column:
            stem_column = column.replace(' 2013', '')
            column_17 = stem_column + ' 2017'
            new_column = 'Differenz ' + stem_column
            output[new_column] = struktur_absolut[column_17] - struktur_absolut[column]
    output['Differenz durchschnittliche Wohndauer'] = struktur_absolut['durchschnittliche Wohndauer 2017'] - struktur_absolut[
        'durchschnittliche Wohndauer 2013']
    output['Differenz SV-Beschäftigtendichte'] = struktur_absolut['SV-Beschäftigtendichte 2017'] - struktur_absolut[
        'SV-Beschäftigtendichte 2013']
    output['Differenz ALG II - Empfängerdichte'] = struktur_absolut[
        'ALG II - Empfängerdichte 2017'] - struktur_absolut['ALG II - Empfängerdichte 2013']
    return output
```

```python
diff = differences(struktur_absolut)
```

Differenz 2020 und 2017

```python
def differences_1720(data):
    output = pd.DataFrame()
    output['Gebietsnummer'] = data['Gebietsnummer']
    for column in data.columns:
        if '2017' in column:
            if column != 'ALG II - Empfängerdichte 2017':
                if column != 'Wahlbeteiligung 2017':
                    stem_column = column.replace(' 2017', '')
                    column_20 = stem_column + ' 2020'
                    new_column = 'Differenz ' + stem_column
                    output[new_column] = struktur_absolut[column_20] - struktur_absolut[column]
    output['Differenz durchschnittliche Wohndauer'] = struktur_absolut['durchschnittliche Wohndauer 2020'] - struktur_absolut[
        'durchschnittliche Wohndauer 2017']
    output['Differenz SV-Beschäftigtendichte'] = struktur_absolut['SV-Beschäftigtendichte 2020'] - struktur_absolut[
        'SV-Beschäftigtendichte 2017']
    return output
```

```python
diff_1720 = differences_1720(struktur_absolut)
```
