```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.colors
import geopandas as gpd
import seaborn as sns
```

```python
pd.set_option('max_columns', None)
sns.set()
```

```python
wahl_diff = pd.read_csv('../Daten/differenzen.csv', index_col = 0)
struktur = pd.read_csv('../Daten/struktur_17_labeled.csv', index_col = 0)
erst_17 = pd.read_csv('../Daten/erststimmen_aufbereitet.csv')
erst_17 = erst_17.merge(struktur, on = 'Gebietsnummer')

wahlbezirke = gpd.read_file('../Daten/wahlbezirke.shp')
wahlbezirke = wahlbezirke.rename(columns = {'Gebietsnum': 'Gebietsnummer'})
stadtbezirke = gpd.read_file('../Daten/stadtteile.shp')

wahl = wahl_diff.merge(struktur, on = 'Gebietsnummer').drop(columns = 'Label')
plot_data = wahlbezirke.set_index('Gebietsnummer').join(wahl_diff.set_index('Gebietsnummer'))
plot_struktur = wahlbezirke.set_index('Gebietsnummer').join(struktur.set_index('Gebietsnummer'))
```

```python
erststimmen = ['Differenz Erststimme CSU', 'Differenz Erststimme SPD', 'Differenz Erststimme GRÜNE', 
              'Differenz Erststimme FDP', 'Differenz Erststimme DIE LINKE', 'Differenz Erststimme AfD']
zweitstimmen = ['Differenz Zweitstimme CSU', 'Differenz Zweitstimme SPD', 'Differenz Zweitstimme GRÜNE', 
              'Differenz Zweitstimme FDP', 'Differenz Zweitstimme DIE LINKE', 'Differenz Zweitstimme AfD']
```

## Verluste SPD

```python
fig, axs = plt.subplots(1, 2, sharex = True, sharey = True, figsize = (15, 15))
plot_data.plot(ax = axs[0], column = 'Differenz Erststimme SPD', legend = True, 
               legend_kwds={'shrink': 0.35}, edgecolor = 'grey')
stadtbezirke.plot(ax = axs[0], facecolor = 'None', edgecolor = 'Black')
axs[0].set_title('Differenz Erststimme SPD')
plot_data.plot(ax = axs[1], column = 'Differenz Zweitstimme SPD', legend = True, 
               legend_kwds={'shrink': 0.35}, edgecolor = 'grey')
stadtbezirke.plot(ax = axs[1], facecolor = 'None', edgecolor = 'Black')
axs[1].set_title('Differenz Zweitstimme SPD')
plt.show()
```

### Korrelation Verluste mit Strukturindikatoren

Verluste der SPD in Erst- und Zweitstimme korrelieren negativ mit Anteil 25-34 Jährigen.

```python
correlation = wahl.drop(columns = 'Gebietsnummer').corr()
value_erst = correlation.iloc[18]['Differenz Erststimme SPD']
value_zweit = correlation.iloc[18]['Differenz Zweitstimme SPD']
def highlight(s):
    is_max = (s == value_erst) | (s == value_zweit)
    return ['background-color: yellow' if v else '' for v in is_max]

correlation.style.apply(highlight)
```

```python
sns.lmplot(x = 'Differenz Erststimme SPD', y = 'Anteil 25 - 34 Jährige Wahlberechtigte 2017', data = wahl)
plt.title('Korrelation Erststimme mit Anteil 25-34 Jährige')
plt.show()
```

```python
sns.lmplot(x = 'Differenz Zweitstimme SPD', y = 'Anteil 25 - 34 Jährige Wahlberechtigte 2017', data = wahl)
plt.title('Korrelation Zweitstimme mit Anteil 25-34 Jährige')
plt.show()
```

## Größte Gewinner in den Wahlbezirken

Wir sehen uns diejenigen Parteien pro Wahlbezirk an, welche Gewinne von über 6% erzielt haben.

```python
# Erststimme
wahl_diff['Gewinner Erststimme'] = 'Kein klarer Gewinner'
wahl_diff.loc[(wahl_diff['Differenz Erststimme FDP'] > 6), 'Gewinner Erststimme'] = 'Gewinn FDP > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme GRÜNE'] > 6), 'Gewinner Erststimme'] = 'Gewinn GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme AfD'] > 6), 'Gewinner Erststimme'] = 'Gewinn AfD > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme FDP'] > 6)
        & (wahl_diff['Differenz Erststimme GRÜNE'] > 6), 'Gewinner Erststimme'] = 'Gewinn FDP und GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme FDP'] > 6)
        & (wahl_diff['Differenz Erststimme AfD'] > 6), 'Gewinner Erststimme'] = 'Gewinn FDP und AfD > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme AfD'] > 6)
        & (wahl_diff['Differenz Erststimme GRÜNE'] > 6), 'Gewinner Erststimme'] = 'Gewinn AfD und GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme AfD'] > 6)
        & (wahl_diff['Differenz Erststimme GRÜNE'] > 6)
        & (wahl_diff['Differenz Erststimme FDP'] > 6), 'Gewinner Erststimme'] = 'Gewinn AfD, FDP und GRÜNE > 6%'

# Zweitstimme
wahl_diff['Gewinner Zweitstimme'] = 'Kein klarer Gewinner'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme FDP'] > 6), 'Gewinner Zweitstimme'] = 'Gewinn FDP > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme GRÜNE'] > 6), 'Gewinner Zweitstimme'] = 'Gewinn GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme AfD'] > 6), 'Gewinner Zweitstimme'] = 'Gewinn AfD > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme FDP'] > 6)
        & (wahl_diff['Differenz Zweitstimme GRÜNE'] > 6), 'Gewinner Zweitstimme'] = 'Gewinn FDP und GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme FDP'] > 6)
        & (wahl_diff['Differenz Zweitstimme AfD'] > 6), 'Gewinner Zweitstimme'] = 'Gewinn FDP und AfD > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme AfD'] > 6)
        & (wahl_diff['Differenz Zweitstimme GRÜNE'] > 6), 'Gewinner Zweitstimme'] = 'Gewinn AfD und GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme AfD'] > 6)
        & (wahl_diff['Differenz Zweitstimme GRÜNE'] > 6)
        & (wahl_diff['Differenz Zweitstimme FDP'] > 6), 'Gewinner Zweitstimme'] = 'Gewinn AfD, FDP und GRÜNE > 6%' 
```

```python
plot_data = wahlbezirke.set_index('Gebietsnummer').join(wahl_diff.set_index('Gebietsnummer'))
cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ['cyan', 'mediumaquamarine', 'hotpink', 'yellow', 'navy', 'sandybrown', 'green', 'lightgray'])
```

```python
fig, axs = plt.subplots(1, 2, sharex = True, sharey = True, figsize = (15, 15))
plot_data.plot(ax = axs[0], column = 'Gewinner Erststimme', legend = True, 
               edgecolor = 'grey', cmap = cmap)
stadtbezirke.plot(ax = axs[0], facecolor = 'None', edgecolor = 'Black')
axs[0].set_title('Gewinner Erststimme')
plot_data.plot(ax = axs[1], column = 'Gewinner Zweitstimme', edgecolor = 'grey', cmap = cmap)
stadtbezirke.plot(ax = axs[1], facecolor = 'None', edgecolor = 'Black')
axs[1].set_title('Gewinner Zweitstimme')
leg = axs[0].get_legend()
leg.set_bbox_to_anchor((2, 0., 0.8, 0.8))
plt.show()
```

### Gewinner in Bezirken mit höchsten SPD-Verlusten

```python
quantil_erst = round(wahl_diff['Differenz Erststimme SPD'].describe()['25%'], 2)
quantil_zweit = round(wahl_diff['Differenz Zweitstimme SPD'].describe()['25%'], 2)

print('25%-Quantil Verluste Erststimme: {}%'.format(quantil_erst))
print('25%-Quantil Verluste Zweitstimme: {}%'.format(quantil_zweit))
```

```python
wahl_diff['Gewinner Erststimme SPD'] = 'Verluste unter threshold'
wahl_diff.loc[(wahl_diff['Differenz Erststimme SPD'] < quantil_erst)
             & (wahl_diff['Differenz Erststimme FDP'] > 6), 'Gewinner Erststimme SPD'] = 'Gewinn FDP > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme SPD'] < quantil_erst)
             & (wahl_diff['Differenz Erststimme GRÜNE'] > 6), 'Gewinner Erststimme SPD'] = 'Gewinn GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme SPD'] < quantil_erst)
             & (wahl_diff['Differenz Erststimme AfD'] > 6), 'Gewinner Erststimme SPD'] = 'Gewinn AfD > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme SPD'] < quantil_erst)
        & (wahl_diff['Differenz Erststimme FDP'] > 6)
        & (wahl_diff['Differenz Erststimme GRÜNE'] > 6), 'Gewinner Erststimme SPD'] = 'Gewinn FDP und GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme SPD'] < quantil_erst)
        & (wahl_diff['Differenz Erststimme FDP'] > 6)
        & (wahl_diff['Differenz Erststimme AfD'] > 6), 'Gewinner Erststimme SPD'] = 'Gewinn FDP und AfD > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme SPD'] < quantil_erst)
        & (wahl_diff['Differenz Erststimme AfD'] > 6)
        & (wahl_diff['Differenz Erststimme GRÜNE'] > 6), 'Gewinner Erststimme SPD'] = 'Gewinn AfD und GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Erststimme SPD'] < quantil_erst)
        & (wahl_diff['Differenz Erststimme AfD'] > 6)
        & (wahl_diff['Differenz Erststimme GRÜNE'] > 6)
        & (wahl_diff['Differenz Erststimme FDP'] > 6), 'Gewinner Erststimme SPD'] = 'Gewinn AfD, FDP und GRÜNE > 6%'
```

```python
wahl_diff['Gewinner Zweitstimme SPD'] = 'Verluste unter threshold'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme SPD'] < quantil_zweit)
             & (wahl_diff['Differenz Zweitstimme FDP'] > 6), 'Gewinner Zweitstimme SPD'] = 'Gewinn FDP > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme SPD'] < quantil_zweit)
             & (wahl_diff['Differenz Zweitstimme GRÜNE'] > 6), 'Gewinner Zweitstimme SPD'] = 'Gewinn GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme SPD'] < quantil_zweit)
             & (wahl_diff['Differenz Zweitstimme AfD'] > 6), 'Gewinner Zweitstimme SPD'] = 'Gewinn AfD > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme SPD'] < quantil_zweit)
        & (wahl_diff['Differenz Zweitstimme FDP'] > 6)
        & (wahl_diff['Differenz Zweitstimme GRÜNE'] > 6), 'Gewinner Zweitstimme SPD'] = 'Gewinn FDP und GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme SPD'] < quantil_zweit)
        & (wahl_diff['Differenz Zweitstimme FDP'] > 6)
        & (wahl_diff['Differenz Zweitstimme AfD'] > 6), 'Gewinner Zweitstimme SPD'] = 'Gewinn FDP und AfD > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme SPD'] < quantil_zweit)
        & (wahl_diff['Differenz Zweitstimme AfD'] > 6)
        & (wahl_diff['Differenz Zweitstimme GRÜNE'] > 6), 'Gewinner Zweitstimme SPD'] = 'Gewinn AfD und GRÜNE > 6%'
wahl_diff.loc[(wahl_diff['Differenz Zweitstimme SPD'] < quantil_zweit)
        & (wahl_diff['Differenz Zweitstimme AfD'] > 6)
        & (wahl_diff['Differenz Zweitstimme GRÜNE'] > 6)
        & (wahl_diff['Differenz Zweitstimme FDP'] > 6), 'Gewinner Zweitstimme SPD'] = 'Gewinn AfD, FDP und GRÜNE > 6%' 
```

```python
plot_data = wahlbezirke.set_index('Gebietsnummer').join(wahl_diff.set_index('Gebietsnummer'))
cmap_1 = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ['cyan', 'mediumaquamarine', 'yellow', 'navy', 'sandybrown', 'green', 'lightgray'])
cmap_2 = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ['cyan', 'yellow', 'navy', 'sandybrown', 'green', 'lightgray'])
```

```python
fig, axs = plt.subplots(1, 2, sharex = True, sharey = True, figsize = (15, 15))
plot_data.plot(ax = axs[0], column = 'Gewinner Erststimme SPD', legend = True, 
               edgecolor = 'grey', cmap = cmap_1)
stadtbezirke.plot(ax = axs[0], facecolor = 'None', edgecolor = 'Black')
axs[0].set_title('Gewinner Erststimme')
plot_data.plot(ax = axs[1], column = 'Gewinner Zweitstimme SPD', edgecolor = 'grey', cmap = cmap_2)
stadtbezirke.plot(ax = axs[1], facecolor = 'None', edgecolor = 'Black')
axs[1].set_title('Gewinner Zweitstimme')
leg = axs[0].get_legend()
leg.set_bbox_to_anchor((2, 0., 0.8, 0.8))
plt.show()
```

Auffällig ist die Präsenz von FDP-Gewinnen, insbesondere in den Bezirken mit den höchsten Verlusten (vgl. oben).

## Interessanter Fall: Studentenstadt

Die Wahlbezirke Studentenstadt (12014) und olympisches Dorf (11025) haben sehr ähnliche Strukturdaten, nur hat Studentenstadt einen etwas höheren Anteil an 18-24 Jährigen (ca. 6 Prozentpunkte). In Studentenstadt hat die SPD jedoch doppelt so hohe Verluste.

```python
wahl_bsp = wahl.loc[wahl['Gebietsnummer'].isin([12014, 11025])]
wahl_bsp.drop(columns = erststimmen + zweitstimmen)
```

Plot der Differenzen von Wahl 2017 und 2013:

```python
cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ['blue', 'red', 'hotpink', 'green', 'yellow', 'violet', 'cyan'])
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 5))
sns.barplot(ax = axs[0], x = wahl_bsp[erststimmen].columns, y = wahl_bsp[erststimmen].iloc[0], 
            palette=['blue', 'red', 'green', 'yellow', 'violet', 'cyan'])
axs[0].set(xticklabels=['CSU', 'SPD', 'GRÜNE', 'FDP', 'DIE LINKE', 'AfD'])
axs[0].set(ylabel = 'Differenz Erststimmen')
axs[0].set(title = 'Olympisches Dorf')
axs[0].set(ylim = (-12.5, 8))
sns.barplot(ax = axs[1], x = wahl_bsp[erststimmen].columns, y = wahl_bsp[erststimmen].iloc[1], 
            palette=['blue', 'red', 'green', 'yellow', 'violet', 'cyan'])
axs[1].set(xticklabels=['CSU', 'SPD', 'GRÜNE', 'FDP', 'DIE LINKE', 'AfD'])
axs[1].set(ylabel = 'Differenz Erststimmen')
axs[1].set(title = 'Studentenstadt')
axs[1].set(ylim = (-12.5, 8))
fig.tight_layout(pad=3.0)
plt.show()
```
