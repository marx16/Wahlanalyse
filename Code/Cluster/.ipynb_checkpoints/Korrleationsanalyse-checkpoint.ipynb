```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
struktur = pd.read_csv('../../../Daten/struktur_17_labeled.csv', index_col = 0)
wahl = pd.read_csv('../../../Daten/differenzen.csv', index_col = 0)
wahl_erst = pd.read_csv('../../../Daten/differenzen_erststimme.csv', index_col = 0)
wahl_zweit = pd.read_csv('../../../Daten/differenzen_zweitstimme.csv', index_col = 0)
wahlbezirke = gpd.read_file('../../../Daten/wahlbezirke.shp')
stadtteile = gpd.read_file('../../../Daten/stadtteile.shp')
wahlbezirke = wahlbezirke.rename(columns = {'Gebietsnum': 'Gebietsnummer'})
```

```python
data = wahl.merge(struktur, on = 'Gebietsnummer')
data_erst = wahl_erst.merge(struktur, on = 'Gebietsnummer')
data_zweit = wahl_zweit.merge(struktur, on = 'Gebietsnummer')
plot_data = wahlbezirke.set_index('Gebietsnummer').join(data.set_index('Gebietsnummer'))
```

### Plot Cluster

Einteilung der Wahlbezirke in Cluster auf Basis der Strukturindikatoren. Angewandte Methoden waren Hauptkomponentenanalyse (PCA) und KMeans Clustering.

```python
fig, ax = plt.subplots()
plot_data.plot(ax = ax, column = 'Label', legend = True, edgecolor = 'grey')
stadtteile.plot(ax = ax, edgecolor = 'Black', facecolor = 'None')
plt.show()
```

### Plot Verluste SPD

Ein Vergleich mit den Verlusten der SPD in den einzelnen Wahlbezirken zeigt, dass die stärksten Verluste größtenteils in Cluster 3 liegen.

```python
fig, ax = plt.subplots()
plot_data.plot(ax = ax, column = 'Differenz Erststimme SPD', legend = True, edgecolor = 'grey', cmap = 'Blues')
stadtteile.plot(ax = ax, edgecolor = 'Black', facecolor = 'None')
plt.show()
```

```python
fig, ax = plt.subplots()
plot_data.plot(ax = ax, column = 'Differenz Zweitstimme SPD', legend = True, edgecolor = 'grey', cmap = 'Blues')
stadtteile.plot(ax = ax, edgecolor = 'Black', facecolor = 'None')
plt.show()
```

## Korrelationen auf Clustern

Korrelationen von Verlusten/Gewinnen der Parteien mit den Strukturindikatoren, gerechnet pro Cluster.

### Cluster 1

Erstimme: Starke negative Korrelation mit CSU und abgeschwächt mit Grünen. Sonst negative Korrelation mit nichtdeutscher Bevölkerung, Migrationshintergrund, Anteil 35 - 44 Jährige und Anteil Einpersonenhaushalte.

Zweitstimme: zusätzlich positive Korrelation mit SV-Beschäftigten, Arbeitslosen und ALG-II Empfängern.

#### Erststimme

```python
label_1 = data_erst.loc[data_erst['Label'] == 1]
label_1.drop(columns = ['Gebietsnummer', 'Label']).corr()
```

#### Zweitstimme

```python
label_1 = data_zweit.loc[data_zweit['Label'] == 1]
label_1.drop(columns = ['Gebietsnummer', 'Label']).corr()
```

### Cluster 2

Erststimme: Starke negative Korrelation mit CSU, abgeschwächt mit Grünen, FDP und Linken. Sonst negative Korrelation mit Anteil 25 - 34 Jährige.

Zweitstimme: zusätzlich negative Korrelation mit Linken.

#### Erststimme

```python
label_2 = data_erst.loc[data_erst['Label'] == 2]
label_2.drop(columns = ['Gebietsnummer', 'Label']).corr()
```

#### Zweitstimme

```python
label_2 = data_zweit.loc[data_zweit['Label'] == 2]
label_2.drop(columns = ['Gebietsnummer', 'Label']).corr()
```

### Cluster 3

Erststimme: Starke negative Korrelation mit CSU und FDP, abgschwächt mit Grünen und leichte positive Korrelation mit Linken. Sonst starke positive Korrelation mit nichtdeutscher Bevölkerung, Migrationshintergrund, Anteil Verheiratete und Arbeitslosendichte, starke negative Korrelation mit Anteil deutscher Bevölkerung, Anteil 45 -  59 Jährige und SV-Beschäftigtendichte.

Zweitstimme: Starke negative Korrelation mit CSU und FDP, abgeschwächt mit Grünen und Linken, positive Korrelation mit AfD. Sonst negative Korrelation mit 25 - 44 Jährigen, postive mit älter als 45 und Verheirateten, abgeschwächt Verheitete, Haushalte mit Kindern, Arbeitslose und ALG-II Empfänger.

#### Erststimme

```python
label_3 = data_erst.loc[data_erst['Label'] == 3]
label_3.drop(columns = ['Gebietsnummer', 'Label']).corr()
```

#### Zweitstimme

```python
label_3 = data_zweit.loc[data_zweit['Label'] == 3]
label_3.drop(columns = ['Gebietsnummer', 'Label']).corr()
```

### Cluster 4

Erststimme: Starke negative Korrelation mit CSU, FDP, abeschwächt mit Grünen. Sonst positive Korrelation mit Nichtdeutschen, Migrationshintergrund und Verheirateten, Arbeitslosen und ALG-II Empfängern. Negative Korrleation mit Deutschen, 45 - 59 Jährigen, Haushalten mit Kindern und SV-Beschäftigtendichte.

Zweitstimme: starke postitive Korrelation mit Grünen und Linken. Sonst starke postitive Korrelation mit Verheirateten und 25 - 44 Jährigen, starke negative Korrelation mit 18 - 24 bzw. 45 - 59 Jährigen und SV-Beschäftigtendichte.
